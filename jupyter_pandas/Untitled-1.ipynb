```python
def tela(tabela, descricao, tipo_botao):
  import PySimpleGUI as sg

  if tipo_botao == 1:
    botao = sg.Ok()
  elif tipo_botao == 2:
    botao = sg.Submit()
  elif tipo_botao == 3:
    botao = sg.Help()

  layout = [ [sg.Text(f"Cadastro de {tabela}", expand_x=True, justification="center")],
             [sg.Text("Campos"), sg.InputText()],
             [sg.Button("Voltar"), sg.Button("Salvar"), botao] ]
  window = sg.Window(tabela, layout)

  while True:
    event, values = window.read()
    if event == sg.WIN_CLOSED or event == "Voltar":
      break
    elif event == "Salvar":
      sg.popup(descricao)
    elif tipo_botao == 1 and event == "Ok":
      if 2 + 2 == 4:
        sg.popup("Funcionou")
    elif tipo_botao == 2 and event == "Submit":
      if 2 * 2 == 4:
        sg.popup("Funcionou")
    elif tipo_botao == 3 and event == "Help":
      if 2 / 2 == 1:
        sg.popup("Funcionou")
  window.close()
```

```python
class Tela():
  def tela(self):
    import PySimpleGUI as sg

    layout = [ [sg.Text(f"Cadastro de {self.tabela}", expand_x=True, justification="center")],
              [sg.Text("Campos"), sg.InputText()],
              [sg.Button("Voltar"), sg.Button("Salvar"), self.botao] ]
    window = sg.Window(self.tabela, layout)

    while True:
      event, values = window.read()
      if event == sg.WIN_CLOSED or event == "Voltar":
        break
      elif event == "Salvar":
        sg.popup(self.descricao)
      
      self.eventoBotao(event)
    window.close()

  def eventoBotao(self, event):
    pass
```

```python
class TelaLaboratorio(Tela):
  def __init__(self):
    self.tabela = "Laboratório"
    self.descricao = "Essa função vai dar trabalho"
    self.botao = sg.Ok()

  def eventoBotao(self, event):
    if event == "Ok":
      if 2 + 2 == 4:
        sg.popup("Funcionou")
```

```python
class TelaMedicamento(Tela):
  def __init__(self):
    self.tabela = "Medicamento"
    self.descricao = "Vixi!!!"
    self.botao = sg.Submit()

  def eventoBotao(self, event):
    if event == "Submit":
      if 2 * 2 == 4:
        sg.popup("Funcionou")
```

```python
class TelaCliente(Tela):
  def __init__(self):
    self.tabela = "Cliente"
    self.descricao = "E agora?"
    self.botao = sg.Help()

  def eventoBotao(self, event):
    if event == "Help":
      if 2 / 2 == 1:
        sg.popup("Funcionou")
```

```python
while True:
  tipo = input("1 - Laboratório; 2 - Medicamento; 3 - Cliente; 4 - Sair: ")

  if tipo == "1":
    TelaLaboratorio().tela()
  elif tipo == "2":
    TelaMedicamento().tela()
  elif tipo == "3":
    TelaCliente().tela()
  else:
    break
```
